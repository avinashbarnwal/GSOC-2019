## 1. Load the shared library

```python
import ctypes
import sys
import numpy as np
from scipy.stats import norm,logistic
import matplotlib.pyplot as plt
%matplotlib inline

if sys.platform == 'darwin':
    libname = './distlib.dylib'
else:
    libname = './distlib.so'

lib = ctypes.cdll.LoadLibrary(libname)
```

```python
# Declare return types
lib.neg_grad.restype = ctypes.c_double
lib.loss.restype     = ctypes.c_double
lib.hessian.restype  = ctypes.c_double
```

## 2. Visualize Loss

```python
# Parameters of distribution
n_obs     = 100
y_obs     = [50,0,20]
x_vals = np.linspace(-50, 50, 200)
```

```python
# Call dnorm() function for every x in x_vals
y_vals_loss = [[lib.loss(ctypes.c_double(n_obs), ctypes.c_double(y_inp), ctypes.c_double(x)) for x in x_vals] for y_inp in y_obs]
# Call dlogis() function for every x in x_vals
```

```python
plt.plot(x_vals, y_vals_loss[0], label='y observed = 50')
plt.plot(x_vals, y_vals_loss[1], label='y observed = 0')
plt.plot(x_vals, y_vals_loss[2], label='y observed = 20')
plt.xlabel('x')
plt.ylabel('Loss')
plt.title('n obs(trials) = 100, Loss for y obs = 50, 0 and 20')
plt.legend(loc='best')
plt.tight_layout()
```

## 3. Visualize Negative Gradient

```python
# Call dnorm() function for every x in x_vals
y_vals_neg_grad = [[lib.neg_grad(ctypes.c_double(n_obs), ctypes.c_double(y_inp), ctypes.c_double(x)) for x in x_vals] for y_inp in y_obs]
# Call dlogis() function for every x in x_vals
```

```python
plt.plot(x_vals, y_vals_neg_grad[0], label='y observed = 50')
plt.plot(x_vals, y_vals_neg_grad[1], label='y observed = 0')
plt.plot(x_vals, y_vals_neg_grad[2], label='y observed = 20')
plt.xlabel('x')
plt.ylabel('Negative Gradient')
plt.title('n obs(trials) = 100, Negative Gradient for y obs = 50, 0 and 20')
plt.legend(loc='best')
plt.tight_layout()
```

## 4. Visualize Hessian

```python
# Call dnorm() function for every x in x_vals
y_vals_hessian = [[lib.hessian(ctypes.c_double(n_obs), ctypes.c_double(y_inp), ctypes.c_double(x)) for x in x_vals] for y_inp in y_obs]
# Call dlogis() function for every x in x_vals
```

## Not Dependent on y obs as Formula doesn't have y obs

```python
plt.plot(x_vals, y_vals_hessian[0], label='y observed = 50')
plt.plot(x_vals, y_vals_hessian[1], label='y observed = 0')
plt.plot(x_vals, y_vals_hessian[2], label='y observed = 20')
plt.xlabel('x')
plt.ylabel('Hessian')
plt.title('n obs(trials) = 100, Hessian for y obs = 50, 0 and 20')
plt.legend(loc='best')
plt.tight_layout()
```
